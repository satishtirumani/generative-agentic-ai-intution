# TF-IDF
TF-IDF is an improvement over Bag of Words that not only counts words but also measures how important a word is in a document.

👉 Idea:

Words that appear frequently in one document but rarely across all documents are more important.

🔹 Why not just Bag of Words?

In BoW, common words like:

“people”, “the”, “is”

get high importance ❌

👉 TF-IDF fixes this by down-weighting common words and boosting rare, meaningful ones.

🔹 Example (simple intuition)
Documents:
1. people watch movie
2. people watch cricket
3. people like movie
"people" → appears in ALL docs → LOW importance
"cricket" → appears in ONE doc → HIGH importance

👉 TF-IDF captures this automatically.

🔹 Key Advantages

✔ Reduces impact of common words
✔ Highlights meaningful terms
✔ Better than BoW for most ML tasks

🔹 Limitations

❌ Still ignores:

Word order
Context/meaning
Synonyms

👉 So TF-IDF is better than BoW, but not as powerful as embeddings.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
tf_idf = TfidfVectorizer()
tf_idf_vector = tf_idf.fit_transform(documents)
tf_idf_vector
tf_idf_vector.toarray()
for vector in tf_idf_vector.toarray():
    print(vector)
tf_idf.get_feature_names_out()
print("Vocabulary:", tf_idf.get_feature_names_out())
```
