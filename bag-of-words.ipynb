# BOW
Bag of Words is a simple NLP technique that converts text into numbers by counting how many times each word appears — ignoring grammar and word order.

👉 Think of it as:

“Just count words, don’t care about sentence structure.”

🔹 Example
Input sentences:
1. "people watch movie"
2. "people watch cricket"
3. "people like movie"
Step 1: Build Vocabulary (unique words)
["people", "watch", "movie", "cricket", "like"]
Step 2: Create vectors (word counts)
Sentence	people	watch	movie	cricket	like
people watch movie	1	1	1	0	0
people watch cricket	1	1	0	1	0
people like movie	1	0	1	0	1

👉 This matrix is your Bag of Words representation

🔹 When to use

✅ Good for:

Spam detection
Simple classifiers
Small NLP tasks

❌ Not ideal for:

Semantic understanding
Chatbots / LLM-level tasks

###### https://scikit-learn.org/0.15/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html

```python
from sklearn.feature_extraction.text import CountVectorizer
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
document=["divesh is a good mentor he teaches us very well he teaches the project"]
bow = CountVectorizer()
bow_matrix =bow .fit_transform(document)
bow_matrix
print("BOW Matrix:\n", bow_matrix.toarray())
bow.get_feature_names_out()
print("Vocabulary:", bow.get_feature_names_out())
bow.fit(documents)
documents
bow.transform(documents).toarray()
document = ["lion is the king of jungle"]
bow.transform(document)
```
